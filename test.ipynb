```python
import pdfplumber

def extract_text_pagewise(pdf_path):
    page_texts = []
    
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text()
            page_texts.append(text if text else "")
    
    return page_texts

# Example usage:
pdf_path = r"C:\Users\abhij\AAROGYA-SPARSHA\data.pdf"
pages = extract_text_pagewise(pdf_path)

for i, content in enumerate(pages):
    # print(f"\n--- Page {i+1} ---\n{content}\n")
    if i+1>=5 and i+1<=19 :
        print(f"{content[8:]}\n")
        break
```

```python
name = "" 
for i in range(len(content)):
    name=name+content[i]
    if content[i]==")":
        break
print(name[8:])

name=name[8:]

```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Folder where images are stored
image_folder = r"C:\Users\abhij\AAROGYA-SPARSHA\images"


# Search for the matching file in the folder
image_file = next(
    (file for file in os.listdir(image_folder) if file.startswith(name) and file.lower().endswith(('.png', '.jpg', '.jpeg'))),
    None
)

if image_file:
    image_path = os.path.join(image_folder, image_file)
    img = mpimg.imread(image_path)

    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(name)
    plt.show()
else:
    print(f"No image found for {name}")
```
